# Stop Words

#### Stop Words; temel olarak gereksiz kelimelerdir. Cumle akisini guzellestirmek icin kullanilan ancak veri analizi yaparken herhangi bir anlam icermeyen kelimelere stop words diyoruz. Ornegin; ingilizcede a, the, an gibi kelimelerdir. Genellikle bir metinde en cok tekrar eden kelimeleri analiz edersek, bu kelimelerin stop words'ler oldugunu gorecegiz. Stop words'ler makineler icin anlam ifade etmediginden cikarilabilirler veya bir esik degeri belirlenip o esik degeri uzerine cikilmaz.


#### NLP'de kullanilan metne 'corpus' diyoruz

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
text = """Lorem Ipsum is simply dummy text of the printing and typesetting industry. Lorem Ipsum has been the industry's standard dummy text ever since the 1500s, when an unknown printer took a galley of type and scrambled it to make a type specimen book. It has survived not only five centuries, but also the leap into electronic typesetting, remaining essentially unchanged. It was popularised in the 1960s with the release of Letraset sheets containing Lorem Ipsum passages, and more recently with desktop publishing software like Aldus PageMaker including versions of Lorem Ipsum."""
```

### Ingilizcede ki stop words'ler.

##### english olan kismi turkish yaparsak turkce icin de gecerli stop words'leri gorebiliriz.

```python
stop_words = stopwords.words('english')
stop_words
```

### Metin uzerinde islem yapmak icin tokenizer yapalim.

```python
word_token = word_tokenize(text)
word_token
```

### Stop Words olanlari metnimizden cikarmak icin filtreleme yapalim.

```python
filtered_words = list(token for token in word_token if token not in stop_words)
filtered_words
```
